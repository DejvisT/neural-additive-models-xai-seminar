#### Training NAM models

Set the `json_path` variable to the json containing the parameter configuration of the model you want to train and run the code to get the training command. Run the training command to train the model.

```python
json_path = "training_conf/nam_training_parameters_housing.json"
```

```python
import json

# Load the JSON file
with open(json_path, "r") as f:
    cfg = json.load(f)

cmd = ["python -m neural_additive_models.nam_train"]

for key, value in cfg.items():
    # null in JSON â†’ None in Python
    if value is None:
        arg = f"--{key}=None"
    # booleans => lowercase true/false for CLI
    elif isinstance(value, bool):
        arg = f"--{key}={'true' if value else 'false'}"
    # strings need no quoting
    else:
        arg = f"--{key}={value}"
    cmd.append(arg)

# Join into one command
final_cmd = " ".join(cmd)
print(final_cmd)
```
